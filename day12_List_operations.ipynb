```python
#list methods
#adding elements to list
#append(ele):adds the elemenst at end of the list
#extand(list):adds all elemetns in list at end
#insert(pos,ele):Insert the element at specified postion
l = [1,2,3,4]
l.append(5)
l.extend([6,7,8,9])
print(l)
l.insert(-1,45)
print(l)
```

```python
#removing element from list
#remove(val): it remove the first occurence from list, otherwise valueError
#pos():It remove and return value from the list,if list is empty it raise error
#pop(postion):It remove and return specified postion element from list
#clear():It removes all elemnts from list
l = [1,2,3,4,5,3]
l.remove(3)
print(l)
val = l.pop()
print(val)
print(l)
print(l.pop(2),l)
#print(l.pop(120), l) #IndexError:pop index out of range
#print([].pop()) #IndexError: pop from empty list
l.clear()
print(l)
```

```python
#Task of Attendance Managemnts
n = int(input("Enter total number of students: "))

present = 0
absent = 0

total = n   # store original value for average

while n > 0:
    val = int(input("Enter 1 for Present, 0 for Absent: "))
    
    if val == 1:
        present += 1
    elif val == 0:
        absent += 1
    else:
        print("Invalid value")
        continue   # ask again without reducing n
    
    n -= 1

print(f"Total present: {present}")
print(f"Total absent: {absent}")

avg = (present / total) * 100
print(f"Attendance percentage: {avg}%")
```
