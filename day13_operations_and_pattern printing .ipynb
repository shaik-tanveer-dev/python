```python
#sparate the even and odd values
s=list(map(int,input().split()))
even=[]
odd=[]
for i in s:
    if i%2==0:
        even.append(i)
    else:
        odd.append(i)
print(even)
print(odd)
```

```python
#find the minumum value
li = list(map(int,input().split()))
min_val=li[0]
for i in range(1,len(li)):
    if min_val>li[i]:
        min_val=li[i]
print(min_val)
```

```python
#find the maxminum value
li = list(map(int,input().split()))
max_val=li[0]
for i in range(1,len(li)):
    if max_val<li[i]:
        max_val=li[i]
print(max_val)
```

```python
#find target element frequency
li = list(map(int,input().split()))
count = 0
target =12
for i in li:
    if i == 12:
        count+=1
print(count)
```

```python
#remove the target elemnt from list
li = list(map(int,input().split()))
target = 3
for i in li:
    if target==i:
        li.remove(target)
print(li)
```

```python
#printing starts
for i in range(5):
    for j in range(5):
        print('*',end=" ")
    print()
```

```python
#printing numbers in colums
for i in range(5):
    for j in range(5):
        print(i+1,end=" ")
    print()
```

```python
#printing numbers rows
for i in range(5):
    for j in range(5):
        print(j+1,end=" ")
    print()
```

```python
'''*
* *
* * * 
* * * *'''
for i in range(5):
    for j in range(5):
        if i >= j:
            print("*",end = " ")
        else:
            print(" " ,end=" ")
    print()
```

```python
#patterns
n=4
for i in range(n):
    for j in range(n):
        if i == 0 or i== n-1 or j==0 or j==n-1:
            print("*",end = " ")
        else:
            print(" ", end=" ")
    print()
```
