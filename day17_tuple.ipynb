```python
#without excetption
try:
    lst=[1,2,3,5]
    print(lst)
    print(lst)
    print(lst[10])
    print('hi')
except:
    print("something went wrong in the program")
else:
    print("program done")
finally:
    print("successfully program excetion done")
```

```python
#with exception
try:
    a=10
    b=0
    print(a/b)
except:
    print("something wrong in program")
else:
    print("program done")
finally:
    print("successfully program excetued successfully")
```

```python
#with exception and handling different error
try:
    a=10
    b=0
    print(a/b)
except ZeroDivisionError:
    print("Zero divison error is not possible")
except AttributeError:
    print("something wrong in the program")
except:
    print("something wrong in program")
else:
    print("program done")
finally:
    print("successfully program excetued successfully")
```
