```python
#Types of attributes
#1.class Attributes
#Class defination
class student:
    school_name = 'ZP High School' #class attribute
    def __init__(self, name, age): #---------------------------->Initilization Constructor
        self.name = name #instance attribute
        self.age = age #instance attribute
    def details(self): #method
        return f"My name is {self.name} and I am a {self.age} years old"
#object Creation

s1 = student("Ravi", 14)
s2 = student("Babu", 13)
print(s1.name)
print(s2.name)
    
```

```python
# class definition 
class Calculator:
    def __init__(self, x: int, y: int):   
        self.x = x
        self.y = y

    def add(self):
        c = self.x + self.y            
        return f"Sum is: {c}"

    def sub(self):
        b = self.x - self.y
        return f"Subtraction is: {b}"

    def mul(self):
        f = self.x * self.y
        return f"Multiplication is: {f}"


# object creation
s1 = Calculator(10, 20)

# method calls
print(s1.add())
print(s1.sub())
print(s1.mul())
```
