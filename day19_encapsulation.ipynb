```python
#encapsulation

class Gameplayer:
    def __init__(self, name):
        self.name = name
        self.__health = 100   # private variable

    def get_health(self):
        return self.__health

    def take_damage(self, damage):
        self.__health -= damage
        if self.__health < 0:
            self.__health = 0
        print(f"{self.name} health: {self.__health}")
        
player = Gameplayer('Hero')
player.take_damage(30)
print(player.get_health())
print(player._Gameplayer__health)
```
