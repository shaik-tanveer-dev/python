```python
#Interitance
class vehicle:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    def start(self):
        print(f"{self.brand} is staring..")
class car(vehicle):
    def __init__(self, brand, year, doors):
        super().__init__(brand, year)
        self.doors = doors
    def horn(self):
        print("Beep..Beep")
class Bike(vehicle):
    def __init__(self, brand, year, type_bike):
        super().__init__(brand, year)
        self.type_bike = type_bike
    def ring_bell(self):
        print("Ring..Ring")

my_car = car("Toyota", 2023, 4)
my_bike = Bike("Hero", 2022, "scooter")
my_car.start()
my_car.horn()
my_bike.start()
my_bike.ring_bell()
```
