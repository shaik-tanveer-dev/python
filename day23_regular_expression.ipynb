```python
#regular expression

import re

string = "The Sky is blue and ocean is blue"
a = re.match("The", string)
print(a)
print(a.group())
print(a.span())
```

```python
#regular expression

import re

string = "The Sky is blue and ocean is blue"
a = re.match("[a-zA-Z]+", string)
print(a)
print(a.group())
print(a.span())
```

```python
#regular expression

import re

string = "The Sky is blue and ocean is blue"
a = re.match(r"[a-z\sA-Z]", string)
print(a)
print(a.group())
print(a.span())
```

```python
#regular expression

import re

string = "The Sky is blue and ocean is blue"
a = re.search("Sky", string)
print(a)
print(a.group())
print(a.span())
```
