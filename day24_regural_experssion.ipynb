```python
#mobile number validation
#consratins:
#starts with 6-9 numbers
#its must be 10 digits
#length should 9
import re

ph = input()
result = re.fullmatch(r"^[6-9][0-9]{9}",ph)
# print(result)
# print(result.span())
# print(result.group())
if result:
    print("valid phn number")
else:
    print("invalid phn number")
```

```python
#Pan number validation
#consratins:
#allows uppercase letters and digits
#first 5 are uppercase
#next 4 digits
#last one is character
import re

pan_number = input()
result = re.fullmatch(r"^[A-Z]{5}[0-9]{4}[A-Z]$",pan_number)
if result:
    print("valid pan number")
else:
    print("Invalid pan number")
```

```python
#Email validation
# constarins:
# It allows lower case letters and digit
# It allows special chaters like [@ .]
# Last domain length 2 or 3

import re

email = input()

result = re.fullmatch(r"^[a-z0-9][a-z0-9.]+\@[a-z0-9]+\.[a-z]{2,3}$",email)
if result:
    print("Valid email")
else:
    print("Invalid email")

```

```python
# Date validation
import re

date = input()

result = re.fullmatch(r"(0?[1-9]|[12][0-9]|3[01])/(0?[1-9]|1[0-2])/(\d{2}|\d{4})", date)

if result:
    print("valid date")
else:
    print("invalid date")
```

```python
#password validation
import re

password = input("Enter password: ")

# Conditions:
# At least 8 characters
# At least 1 lowercase, uppercase, digit, special char

pattern = r'^(?=.*[a-z])(?=.*[A-Z])(?=.*\d)(?=.*[@$!%*?&]).{8,}$'

if re.fullmatch(pattern, password):
    print("Strong password")
else:
    print("Weak password")
```
