```python
from datetime import datetime, timedelta

current_datetime = datetime.now()
print(current_datetime)
today = current_datetime.date()
print(today)
curr_time = current_datetime.time()
print(curr_time)
```

```python
#getting year, month and day values from today date
print(today.year)
print(today.month)
print(today.day)
```

```python
#getting hour, minute, second from the curren_datetime
print(current_datetime)
print(current_datetime.hour)
print(current_datetime.minute)
print(current_datetime.second)
print(current_datetime.microsecond)
```

```python
# max it returns maximum value in that attribute
print(datetime.max)
print(today.max)
print(curr_time.max)

# min it returns minimum value in that attribute
print(datetime.min)
print(today.min)
print(curr_time.min)
```

```python
#replace
print(today)
print(f"modified date {today.replace(year = 2025, month = 11)}")
```

```python
str_date = today.strftime("%d - %m - %y")
print(str_date)
print(str_date[:5])
print(today.strftime("%d - %B - %Y"))
```

```python
print(today + timedelta(days=10))
print(today - timedelta(days =20))
```

```python
# clock implemention

from IPython.display import clear_output

sec_counter = 0
mint_counter = 0
hours_counter = 0
while True:
    print(f"{hours_counter} : {mint_counter} : {sec_counter}")
    time.sleep(1)
    sec_counter += 1
    if sec_counter == 60:
        mint_counter += 1
        sec_counter = 0
        if mint_counter == 60:
            hours_counter += 1
            mint_counter = 0
            if hours_counter == 24:
                hours_counter = 0
    clear_output(wait = True)
```
