```python
# importing required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Housing.csv')
print(df)

df.info() #check the cloum type
```

```python
map_values = {'no':0, 'yes':1}
df['mainroad'] = df['mainroad'].replace(map_values)
df.head()
```

```python
map_values = {'no':0,'yes':1}
df['guestroom'] = df['guestroom'].replace(map_values)
df['basement'] = df['basement'].replace(map_values)
df['hotwaterheating'] = df['hotwaterheating'].replace(map_values)
df['airconditioning'] = df['airconditioning'].replace(map_values)
df['prefarea'] = df['prefarea'].replace(map_values)
df.head()
```

```python
df['furnishingstatus']=df['furnishingstatus'].replace({'furnished':2, 'semi-furnished':1, 'unfurnished':0})
```

```python
df.mainroad=df.mainroad.astype('int')
df.airconditioning=df.airconditioning.astype('int')
df.basement=df.basement.astype('int')
df.guestroom=df.guestroom.astype('int')
df.prefarea=df.prefarea.astype('int')
df.furnishingstatus=df.furnishingstatus.astype('int')
df.hotwaterheating=df.hotwaterheating.astype('int')
df.info()
```

```python
# Data Visualization
plt.Figure(figsize=(20,20))
sns.heatmap(data=df,annot=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=df[['bedrooms', 'bathrooms', 'stories', 'mainroad',
       'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
       'parking', 'prefarea', 'furnishingstatus']])
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.barplot(data=df[['price','area']])
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df,x='area',y='price',hue='mainroad')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data= df.loc[100:200],x='area',y='price',hue='mainroad')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=df,x='area',y='price',hue='mainroad')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(data=df,x='area',y='price',hue='mainroad')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.kdeplot(data=df,x='area')
plt.grid(True)
plt.show()
```

```python
sns.countplot(data=df, x = 'mainroad', hue = 'parking')
plt.show()
```
